```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np
import matplotlib.pyplot as plt
# Synthetic sensor data setup
n = 200
mean_a, std_a = 25.0, 3.0
mean_b, std_b = 27.0, 4.5

# Replace 1234 with the last 4 digits of your Drexel ID
seed = 7889
rng = np.random.default_rng(seed)

# Simulated temperature readings (deg C)
sensor_a = rng.normal(loc=mean_a, scale=std_a, size=n)
sensor_b = rng.normal(loc=mean_b, scale=std_b, size=n)

# Uniform timestamps in [0, 10] seconds, sorted for natural time order
timestamp = np.sort(rng.uniform(0, 10, size=n))
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
plt.figure(figsize=(9, 5))
plt.scatter(timestamp, sensor_a, color='tab:blue', alpha=0.75, s=24, label='Sensor A')
plt.scatter(timestamp, sensor_b, color='tab:orange', alpha=0.75, s=24, label='Sensor B')

plt.xlabel('Timestamp (s)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor Temperature Readings vs Time')
plt.legend()
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.

plt.figure(figsize=(10, 6))
plt.hist(sensor_a, bins=30, alpha=0.5, color='tab:blue', label='Sensor A', edgecolor='black', linewidth=0.5)
plt.hist(sensor_b, bins=30, alpha=0.5, color='tab:orange', label='Sensor B', edgecolor='black', linewidth=0.5)

# Add vertical dashed lines at each sensor's mean
plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label=f'Sensor A mean: {mean_a:.1f}°C')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label=f'Sensor B mean: {mean_b:.1f}°C')

plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution Comparison: Sensor A vs Sensor B')
plt.legend()
plt.grid(alpha=0.25, axis='y')
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
# Calculate overall mean of both sensors combined
overall_mean = np.mean([sensor_a, sensor_b])

plt.figure(figsize=(8, 6))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])

# Add horizontal dashed line at overall mean
plt.axhline(overall_mean, color='black', linestyle='--', linewidth=2, label=f'Overall Mean: {overall_mean:.2f}°C')

plt.xlabel('Sensor')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Distribution Comparison: Box Plot')
plt.legend()
plt.grid(alpha=0.25, axis='y')
plt.tight_layout()
plt.show()
```
